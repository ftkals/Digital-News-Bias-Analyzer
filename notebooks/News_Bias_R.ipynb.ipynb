LINGUISTIC ANALYSIS

```r
# Make all headlines lowercase (using tolower)
headline_data$Headline_Text <- tolower(headline_data$Headline_Text)
```

```r
# Create an empty column to store our results
headline_data$Tone <- ""
```

```r
# This loop iterates through each row of the 'headline_data' dataframe.
for (i in 1:nrow(headline_data)) {

  # It checks if the 'Headline_Text' in the current row 'i' contains any of the specified negative words ('crisis', 'stall', 'decline').
  if (grepl("crisis|stall|decline", headline_data$Headline_Text[i])) {
    # If a negative word is found, the 'Tone' for that headline is set to 'Negative'.
    headline_data$Tone[i] <- "Negative"
  } else {
    # Otherwise (if no negative words are found), the 'Tone' is set to 'Positive'.
    headline_data$Tone[i] <- "Positive"
  }

}
#grepl()- grep logical: R function that searches for matches to a regular expression (pattern) within a character vector. It returns a logical vector (TRUE/FALSE) indicating whether each element of the character vector contains a match.
```

```r
# Convert it to a Factor
headline_data$Tone <- factor(headline_data$Tone)
headline_data$Tone
```

```r
# Count the number of headlines for each tone
tone_counts <- table(headline_data$Tone)
#table() is an R function that builds a frequency table of categorical variables. It counts the occurrences of each unique value (or combination of values) in one or more vectors (or factors).

# Create a bar plot
barplot(tone_counts,
        col = c("red", "lightgreen"),
        main = "Distribution of Headline Tone (Positive vs. Negative)",
        xlab = "Tone",
        ylab = "Number of Headlines")
```
